This notebook gives a short introduction into how integration is handled in
`skactiveml`.

```python
# TODO: maybe add expect: func, distribution -> E[func(x)], x ~ distribution
import matplotlib.pyplot as plt
import numpy as np
import time
from scipy.stats import norm

from skactiveml.regressor.estimator._nichke import NormalInverseChiKernelEstimator
from skactiveml.utils._approximation import conditional_expect
from skactiveml.visualization._data_set import gaussian_noise_generator_1d
```

Many query strategies in the field of active learning for regression rely on
numerical integration. Usually it is interesting to know the expected value
of a utility function if a sample x has the target value y. For this expected
value a closed form solution can not always be obtained, hence numerical integration
is necessary. Thus, the problem of calculating $\mathbb{E}[f(y)]$, where $y \sim est(x)$
for a utility function appears. The different ways in which this expected value
can be calculated numerically are handled by the function `conditional_expect`.

To explain this method, consider the following situation: Our utility function at
a given point is given by `f`. We are given the samples `X` with labels `y`,
which we use to estimate the conditional probability function `cond_est`.

```python
random_state = 0
n_samples = 100
x_low, x_up = 0, 2

def f(X):
    return - 1/10*X**3 + X**2 + 8*X + 16

def noise_gen(X):
    return gaussian_noise_generator_1d(X, (0, 1, 4), (0, 1, 1), random_state=random_state)

X = np.linspace(x_low, x_up, num=200).reshape(-1, 1)
y = noise_gen(X)
cond_est = NormalInverseChiKernelEstimator(metric_dict={'gamma': 10.0})
cond_est.fit(X, y)

_, ax = plt.subplots()
mean, std = cond_est.predict(X, return_std=True)
ax.scatter(X, y)
ax.plot(X, mean)
ax.plot(X, mean + std)
ax.plot(X, mean - std)
```

We now want to know at each point in the feature space, what the expected utility is.
For this specific function a closed form solution can be obtained which is not possible
in general.

```python
val_exact = np.zeros(len(X))
for idx, x in enumerate(X):
    cond_dist = cond_est.estimate_conditional_distribution([x])
    val_exact[idx] = -1/10*cond_dist.moment(3) + cond_dist.moment(2) + 8*cond_dist.moment(1) + 16

_, ax = plt.subplots()
ax.scatter(X, val_exact)
ax.set_title('analytical solution')
```

To calculate a numerical approximation we will use the function `conditional_expect`,
which provides different approximation methods.

Some of these methods rely on certain assumptions, which determine how well
the integration method works. For this it is useful to see how the utility function
behaves in the regions where the conditional probability is high.

```python
_, ax = plt.subplots()
y_space = np.linspace(-10, 10, 100)
ax.plot(y_space, norm.pdf(y_space, scale=4))
ax.plot(y_space, norm.pdf(y_space, scale=1))
ax_t = ax.twinx()
ax_t.plot(y_space, f(y_space))
```

For example the method `assume_linear` only works well if the function is approximately linear
to a high probability. To get a feeling for how well different methods work under which
assumptions and how efficient they are, we specify different methods as parameters.

```python
parameters = [
    ("linear", {"method":"assume_linear"}),
    ("exact", {"method":"quad"}),
    ("mc 10", {"method":"monte_carlo", "n_integration_samples":10}),
    ("mc 100", {"method":"monte_carlo", "n_integration_samples":100}),
    ("simpson 3", {"method":"quantile", "quantile_method":"simpson", "n_integration_samples":3}),
    ("simpson 10", {"method":"quantile", "quantile_method":"simpson", "n_integration_samples":10}),
    ("simpson 100", {"method":"quantile", "quantile_method":"simpson", "n_integration_samples":100}),
]

_, axes = plt.subplots(1, len(parameters), figsize=(20, 5), sharey='all')
for (title, parameter), ax in zip(parameters, axes):
    t0 = time.time()
    val = conditional_expect(X, f, cond_est, **parameter, vector_func='optional', random_state=random_state)
    t1 = time.time()
    dt = t1 - t0
    acc = np.average(np.abs(val_exact-val))
    ax.scatter(X, val)
    ax.set_title(title, fontdict={'fontsize': 15, 'color': 'white'})
    ax.set_xlabel(f'time: {dt} \n acc: {acc}', fontdict={'color': 'white'})
```

The plot to left with caption `linear` calculates the expected utility at a point
in the feature space by simpling taking the function value at the expected target
value for that point in the feature space. It can be seen that this works well
on the right side of the feature space. There the standard deviation is much smaller
and thus the function is much more linear to a higher probability.

The second plot from the left with caption `exact` uses the method `expect`
from `scipy`. This method uses an adaptive version of Gaussian quadrature
to calculate the integral from $-\infty$ to $\infty$ of $f(y)est(y|x)$. This
method is exact but takes long. Further specifications can be
passed via `quad_dict`.

The third and forth plot from the left with caption `mc 10` and `mc 100` use
basic monte carlo integration.

The fifth, sixth and seventh plot from the left with caption `simpson 3`, `simpson 10`
and `simpson 100` calculate the integral from $0$ to $1$ of $f(quad(p))$ using `simpson`
integration with 3, 10 or 100 evenly spaced points. Here $quad(p)$ means the quantile function
of $est(y|x)$.
